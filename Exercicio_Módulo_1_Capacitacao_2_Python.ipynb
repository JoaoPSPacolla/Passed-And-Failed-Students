# OBS: SE QUISER TESTAR ESSE CÓDIGO, É PRECISO IMPORTAR A BASE DE DADOS

```python
import pandas as pd
```

```python
df = pd.read_excel('alunos_notas.xlsx')
df.head(6)
```

Aluno mais velho

```python
aluno_mais_velho = df.loc[df['IDADE'].idxmax()]
print(aluno_mais_velho)
```

Aluno mais novo

```python
aluno_mais_novo = df.loc[df['IDADE'].idxmin()]
print(aluno_mais_novo)
```

```python
aprovados = df[df['MÉDIA'] > 5]
print(aprovados)
```

```python
aprovados_por_turma = aprovados.groupby('TURMA').size()
print(aprovados_por_turma)
```

```python
reprovados = df[df['MÉDIA'] <= 5]

reprovados.tail() #Olhando só os ultimos 5 alunos reprovados
```

```python
reprovados_por_turma = reprovados.groupby('TURMA').size()
print(reprovados_por_turma)
```

# Bonus - Analisando especificamente alunos de medicina

```python
reprovados_Medicina = df[(df['TURMA'] == "MEDICINA") & (df['MÉDIA'] <= 5)] #Poderia ter usado a variavel reprovados, mas quis deizar assim para também verem que funciona

reprovados_Medicina
```

```python
total = reprovados_Medicina.shape[0] #zero conta o numero de linhas (1 conta colunas)
print(f"Total de alunos reprovados na turma de Medicina: {total}")
```
